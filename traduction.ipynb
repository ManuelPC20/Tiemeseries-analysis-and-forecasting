```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
df = pd.read_csv("Data/actes-criminels.csv")
df.head()
```

```python
def translate_category(cat):
    translation_list = [
    ['Vol de véhicule à moteur', 'Méfait', 'Vols qualifiés', 'Vol dans / sur véhicule à moteur', 'Introduction', 'Infractions entrainant la mort'],
    ['Motor vehicle theft', 'Mischief', 'Robbery', 'Theft from/into a motor vehicle', 'Breaking and entering', 'Offences resulting in death']
    ]
    if cat in translation_list[0]:
        index = translation_list[0].index(cat)
        return translation_list[1][index]
    

def translate_quart(quart):
    translation_list = [
    ['jour', 'soir', 'nuit'],
    ['Day', 'Evening', 'Night']
    ]
    if quart in translation_list[0]:
        index = translation_list[0].index(quart)
        return translation_list[1][index]
```

```python
df["CATEGORIE"] = df["CATEGORIE"].apply(translate_category)
df["QUART"] = df["QUART"].apply(translate_quart)
```

```python
def get_neighbour(coords):
    locator = Nominatim(user_agent='openmapquest')
    location = locator.reverse(coords)
    try:
        return location.raw["address"]["neighbourhood"]
    except KeyError:
        try:
            return location.raw["address"]["town"]
        except:
            return location.raw["address"]["suburb"]
```
